# Making the First Subset
Now that the file has successfully been uploaded, we can now separate the data column titled "Dog_Breed" to just show the specific dog breed titled "Beagle".
To do this, enter the filtering command below:

        df[df["Dog_Breed"] == "Beagle"]
This should return *true* for all dogs that have 'Beagle' for their listed breed.

After this, you will need to assign a name to this new subset. To do this, enter the new code seen below:
        
        Beagle_subset = df[df["Dog_Breed"] == "Beagle"].copy()
        
-The .copy() is very important to include as it helps avoid error messages later. Now, you should be able to run your data subset by entering "Beagle_subset".

**Congratulations!** You have now completed the creation of your first subset.
